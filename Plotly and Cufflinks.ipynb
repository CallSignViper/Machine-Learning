```python
cardano = pd.read_excel('Cardano.xlsm',sheetname='Source')
```

```python
cardano.head()
```

```python
cardano.iplot(kind='line',x='Date',y='Close**')
```

```python
cardano.iplot(kind='scatter',x='Date',y='Close**',mode='markers')
```

```python
cardano.iplot(kind='bar',x='Date',y='Volume')
```

```python
cc=cardano.corr()
```

```python
sns.heatmap(cc,annot=True,cmap='coolwarm')
```

```python
cardano.iplot(kind='box')
```

```python
spread=cardano[['Close**','Open*']]
```

```python
spread.iplot(kind='surface')
```

```python
spread['Open*'].iplot(kind='hist')
```

```python
spread.iplot(kind='spread')
```

```python
spread.iplot(kind='bubble',x='Open*',y='Close**',size=3)
```
