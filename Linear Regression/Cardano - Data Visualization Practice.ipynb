```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
cardano = pd.read_excel('Cardano.xlsm',sheetname='Source')
```

```python
cardano.set_index('Date')
```

```python
cardano['Close**'].plot.line()
```

```python
x = cardano['Date']
y = cardano['Close**']
```

```python
plt.plot(x,y,'r-')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Cordano')
```

```python
fig = plt.figure()

#list passed in args = left bottom width height
axes = fig.add_axes([1,1.2,1.2,1])

axes.plot(x,y)
axes.set_xlabel('Date')
axes.set_ylabel('Close Price')
axes.set_title('Closing Price by Month')
```

```python
v = cardano['Volume']
fig = plt.figure()

axes1 = fig.add_axes([1,1,1,1])
axes2 = fig.add_axes([1,.5,1,.33])

axes1.plot(x,y)
axes1.set_title('Price')
#axes1.set_xlim(['4/01/2018','04/5/2018'])

axes2.plot(v)
axes2.set_title('Volume')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('Price')

axes[1].plot(v,y)
axes[1].set_title('Volume')

plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,label='Price')
ax.plot(x,v,label='Volume',color='green')

ax.legend(loc=0)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y,color='red',linewidth=1,alpha=1,linestyle='-',marker='*',markersize=10,markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green')

axes[1].plot(x,v)

plt.tight_layout()
```

```python
fig.savefig('Cordano Price and Volume History.pdf',dpi=200)
```
